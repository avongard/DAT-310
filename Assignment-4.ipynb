```python
# Shrink of data to get only what we need
# There are no null values
# Year min = 1969, year max = 2016
# Engine types = Petrol, diesel, gas, other
selected_columns = ['Year', 'Engine Type']
engine_year = data[selected_columns]

# Count the values for each engine type and its corresponding year
engine = engine_year.groupby(['Engine Type', 'Year'])['Engine Type'].count()
engine_type_counts = engine.reset_index(name='Count')
print(engine_type_counts)
```

```python
# Get unique engine types
engine_types = engine_year['Engine Type'].unique()

# Color types for loop
colors = ['red', 'green', 'blue', 'yellow']

# Plotting
plt.figure(figsize=(10, 6))

# Loops through each unique type/color and plots the graph
for engine_type, color in zip(engine_types, colors):
    engine_data = engine_type_counts[engine_type_counts['Engine Type'] == engine_type]
    plt.plot(engine_data['Year'], engine_data['Count'], label=engine_type, color=color)

# Creates labels, title, adds a legends, and displays the graph
plt.xlabel('Year')
plt.ylabel('# of Cars')
plt.title('Car Engine Types Across the Years')
plt.legend()  
plt.show()
```

```python
# Shrink of data to get only what we need
# There are no null values
# Year min = 1969, year max = 2016
# Body = Sedan, van, crossover, wagon, hatch, other
selected_columns = ['Year', 'Body']
body_year = data[selected_columns]

# Count the values for each engine type and its corresponding year
body = body_year.groupby(['Body', 'Year'])['Body'].count()
body_type_counts = body.reset_index(name='Count')
print(body_type_counts)
```

```python
# Get unique body types
body_types = body_year['Body'].unique()

# Color types for loop
colors = ['red', 'green', 'blue', 'yellow', 'cyan', 'magenta']


# Loops through each unique type/color and plots the graph
for body_type, color in zip(body_types, colors):
    body_data = body_type_counts[body_type_counts['Body'] == body_type]
    plt.plot(body_data['Year'], body_data['Count'], label=body_type, color=color)

# Creates labels, title, adds a legends, and displays the graph
plt.xlabel('Year')
plt.ylabel('# of Cars')
plt.title('Body Types Across the Years')
plt.legend() 
plt.show()
```

```python
# Shrink of data to get only what we need
# There are no null values
# Year min = 1969, year max = 2016
# Brand = BMW, Benz, Audi, Toyota, Renault, Volkswagen, Mitsubishi
selected_columns = ['Year', 'Brand']
brand_year = data[selected_columns]

# Count the values for each engine type and its corresponding year
brand = brand_year.groupby(['Brand', 'Year'])['Brand'].count()
brand_type_counts = brand.reset_index(name='Count')
```

```python
# Get unique brand types
brand_types = brand_year['Brand'].unique()

# Color types for loop
colors = ['red', 'green', 'blue', 'yellow', 'cyan', 'magenta', 'black']


# Loops through each unique type/color and plots the graph
for brand_type, color in zip(brand_types, colors):
    brand_data = brand_type_counts[brand_type_counts['Brand'] == brand_type]
    plt.plot(brand_data['Year'], brand_data['Count'], label=brand_type, color=color)

# Creates labels, title, adds a legends, and displays the graph
plt.xlabel('Year')
plt.ylabel('# of Cars')
plt.title('Brand Types Across the Years')
plt.legend() 
plt.show()
```
